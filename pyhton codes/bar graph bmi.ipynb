```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your BMI data (replace with the path to your actual file)
bmi_data = pd.read_csv(r"C:\Users\bgane\Desktop\TOPICS DATA\obese bmi data.csv")

# Filter data for 2022 and calculate the total (sum) by region
data_2022 = bmi_data[bmi_data['Period'] == 2022].groupby('ParentLocation')['FactValueNumericLow'].sum().reset_index()

# Create a bar chart for regional comparison in 2022
plt.figure(figsize=(10, 6))
plt.bar(data_2022['ParentLocation'], data_2022['FactValueNumericLow'])

# Customize the chart
plt.title('Obesity Prevalence Totals by Region (2022)', fontsize=14)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Obesity Prevalence (%)', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()



```
